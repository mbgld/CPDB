```python
%matplotlib inline
import pandas as pd
import seaborn as sns; sns.set()
import numpy as np
from matplotlib import pyplot as plt
from scattermap import scattermap
```

```python
df_1 = pd.read_csv('./out/pvalues.txt', sep='\t', index_col='interacting_pair')
df_1.head()
```

```python
df_1.columns.tolist()
```

```python
df_2=df_1[['secreted', 'is_integrin',
            'Cancer|Alveolar type I', 'Cancer|Alveolar type II', 'Cancer|Basal',
            'Cancer|Cancer', 'Cancer|Ciliated', 'Cancer|Club', 'Cancer|Neuroendocrine']]
df_2.shape
```

```python
df_3 = df_2[((df_2['secreted']==True) & (df_2['is_integrin']==False))]
df_3.shape
```

```python
df_4 = df_3[(df_3['Cancer|Alveolar type I']<0.001)|
            (df_3['Cancer|Alveolar type II']<0.001)|
            (df_3['Cancer|Basal']<0.001)|            
            (df_3['Cancer|Cancer']<0.001)|
            (df_3['Cancer|Ciliated']<0.001)|
            (df_3['Cancer|Club']<0.001)|
            (df_3['Cancer|Neuroendocrine']<0.001)]
print(df_4.shape, df_4.ndim)
df_4
```

```python
df_4.columns.tolist()
```

```python
df_5 = df_4[[ 'Cancer|Alveolar type I',
              'Cancer|Alveolar type II',
              'Cancer|Basal',
              'Cancer|Ciliated',
              'Cancer|Club',
              'Cancer|Neuroendocrine']]
df_6 = df_5.sort_index(ascending=True)
```

```python
df_6.to_csv('./CA_EP.csv')
```

```python
fig, ax = plt.subplots(figsize=(20, 35))

plt.title("Cancer_Epitheial interaction: Threshold=0.33", size=24)
plt.rc('xtick', labelsize=12) # fontsize of the tick labels 
plt.rc('ytick', labelsize=12) # fontsize of the tick labels 
plt.rc('legend', fontsize=12)

ax=scattermap(df_6, marker_size=300, vmin=0, vmax=1, cmap='afmhot') #cmap=plt.cm.RdYlGn
plt.savefig("CA_EP.jpg",dpi=1200)
```
